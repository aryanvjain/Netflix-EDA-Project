## removing cast and description
the cast and the description columns were removed as they were not required for this specific analysis as they dont answer any questions and dont help in filtering/searching too

## dropping the exactly duplicated rows as its the  same thing that occurs  twice

## removing missing value rows
we checked  using isnull().sum() that duration and rating  have missing values  and removing them doesnt afffect the data set as they are very less but help in better analysis

```python
df=df.dropna(subset=['duration','rating'])
```

## new columns
new  columns such as year and  month added were added to the dataset  as they allow to understand additions over years and months  easily instead of dates, this is feature enginnering

```python
df['date_added']=pd.to_datetime(df['date_added'],errors='coerce')
```

```python
df['listed_in']=df['listed_in'].str.title()
```

```python
df['title']=df['title'].str.title()
```

```python
df['year_added']=df['date_added'].dt.year
df['month_added']=df['date_added'].dt.month
```

```python
df['director'].nunique()
```

## drop director
the director column was dropped as there are more  than 5000 directors  in 8000  rows which woould not help  us answer any questions

```python
df.drop(columns=['director'],inplace=True)
```

### 2) country distribution

```python
country_counts=(df['country'].str.split(", ").explode().value_counts())
```

```python
top20_countires=country_counts.head(20)
```

```python
others=country_counts[20:].sum()
```

```python
top20_countires['Others']=others
```

```python
top20_countires=top20_countires.sort_values()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(17,6))
top20_countires.plot(kind='barh')
plt.title('top 20 countries and others on netflix')
plt.xlabel('number of releases')
plt.ylabel('country names')
plt.savefig("top20_countries.png", dpi=300, bbox_inches="tight")
plt.show()
```

### observation
the most popular country is United states by a large margin

### understanding
the possible reaasons for this are:-
1) people of US  are more into entertainment than other countries
2) Netflix focuses more on hollywood

### 3) top 3 years with highest releases

```python
top10years=df['year_added'].value_counts().iloc[:10]
```

```python
top10years['others']=df['year_added'].value_counts().iloc[10:].sum()
```

```python
top10years
```

```python
plt.figure(figsize=(17,6))
top10years.plot(kind='bar')
plt.title('top 10 years with most releases')
plt.xlabel('years')
plt.ylabel('number of releases')
plt.savefig("top_years", dpi=300, bbox_inches="tight")
plt.show()
```

### observation

### understanding

### 4) whats trending last 10 years (tv shows/ movies)

```python
plt.figure(figsize=(27,10))
trending_lately=df.groupby(['year_added','type']).size().unstack(fill_value=0)
trending_lately.plot(kind='bar')
plt.title('whats trending lately')
plt.xlabel('years')
plt.ylabel('number of releases')
plt.savefig("movies_vs_tvshows.png", dpi=300, bbox_inches="tight")
plt.show()
```

### observation
movies have always been added more

### understanding
the possible reaasons for this are:-
1) people prefer movies over tv shows as they are shorter

### 8) duration distribution of movies

```python
movies = df[df["type"] == "Movie"].copy()
movies = movies.dropna(subset=["duration"])

movies["duration"] = movies["duration"].str.replace(" min", "").astype(int)
plt.hist(movies["duration"], bins=30)
plt.title("Duration Distribution of Movies")
plt.xlabel("Duration (minutes)")
plt.ylabel("Number of Movies")
plt.savefig("duration_distribution_of_movies.png", dpi=300, bbox_inches="tight")
plt.show()
```

### observation
movies are made with avg duration between 90-120 mins

### understanding
the possible reaasons for this are:-
1) directors feel this is the sweetspot for viewers, not tooo long and not too short

### 9) avg movie duration

```python
avg_duration = movies["duration"].mean()

print(f"Average Movie Duration: {avg_duration:.2f} minutes")
```

```python
avg_movies= df['year_added'].value_counts().mean()
print(f"Average Movie released : {avg_movies:.2f} ")
```

# Conclusion

This project gave me hands-on experience with the complete Exploratory Data Analysis (EDA) process using the Netflix Titles dataset. Starting with raw data, I cleaned the dataset by handling missing values, removing unnecessary columns, standardizing text, converting data types, and creating new features like **Year Added** and **Month Added** to make the analysis more meaningful.

Through this analysis, I found that Movies make up a much larger portion of Netflix's catalog than TV Shows. The United States and India contribute the highest number of titles, while genres like **International Movies**, **Dramas**, and **Comedies** are among the most common. I also observed that Netflix added the highest number of titles during **July** and **December**. Although this could be related to holiday seasons when people generally have more free time, the dataset doesn't contain enough information to confirm this. Most movies also fall within the standard feature-length duration, averaging around **100 minutes**.

One of the biggest lessons from this project was understanding the importance of making conclusions based only on what the data actually tells us. Since this dataset contains only Netflix's content catalog and not user viewing data, I couldn't make claims about what viewers prefer or what content is the most popular. Instead, I focused on identifying patterns and supporting them with evidence from the data.

Overall, this project helped me become more comfortable with **Pandas**, **Matplotlib**, and **Seaborn**, while also improving my understanding of data cleaning, feature engineering, visualization, and drawing meaningful insights from real-world data. As my first end-to-end EDA project, it has given me a strong foundation for more advanced data analysis and machine learning projects in the future.
